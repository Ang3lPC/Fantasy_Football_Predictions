```python
from bs4 import BeautifulSoup
import requests
from urllib.request import Request, urlopen
import pandas as pd
from sklearn.model_selection import train_test_split, cross_validate, cross_val_score, GridSearchCV,\
RandomizedSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error, mean_absolute_error
from operator import itemgetter
import time

import seaborn as sns
from matplotlib import pyplot as plt
```

```python
offense_nfl_df = pd.read_csv('offense_nfl_2022.csv')
```

```python
offense_nfl_df.info()
```

```python
test_df = offense_nfl_df[['Date','Team','Opp1','Pass_Yds','Pass_TD','Int','Date.1','Team.1','Opp2','Rush_Yds','Rush_TD','Date.2','Team.2','Opp3','Rec','Rec_Yds','Rec_TD','Tar']]
```

```python
# Rows with 0 information, so these will be dropped as they don't add value
test_df.loc[(test_df['Opp1'] == '0') & (test_df['Opp2'] == '0') & (test_df['Opp3'] == '0')]
```

```python
clean_nfl_df = test_df.loc[(test_df['Opp1'] != '0') | (test_df['Opp2'] != '0') | (test_df['Opp3'] != '0')]
```

```python
clean_nfl_df.loc[clean_nfl_df['Opp1'] == "0", "Opp1"] = clean_nfl_df['Opp3']
clean_nfl_df.loc[clean_nfl_df['Opp1'] == "0", "Opp1"] = clean_nfl_df['Opp2']
```

```python
clean_nfl_df.loc[clean_nfl_df['Opp1'] != "0"]
```

```python
clean_nfl_df.loc[clean_nfl_df['Team'] == "0", "Team"] = clean_nfl_df['Team.2']
clean_nfl_df.loc[clean_nfl_df['Team'] == "0", "Team"] = clean_nfl_df['Team.1']
```

```python
clean_nfl_df.loc[clean_nfl_df['Date'] == "0", "Date"] = clean_nfl_df['Date.2']
clean_nfl_df.loc[clean_nfl_df['Date'] == "0", "Date"] = clean_nfl_df['Date.1']
```

```python
clean_nfl_df = clean_nfl_df[['Date','Team','Opp1','Pass_Yds','Pass_TD','Int','Rush_Yds','Rush_TD','Rec','Rec_Yds','Rec_TD','Tar']]
```

```python
# Changing date to datetime in order to filter only regular season games.
clean_nfl_df['Date'] = pd.to_datetime(clean_nfl_df['Date'])
```

```python
clean_nfl_df
```

```python
# September 8, 2022 – January 8, 2023
clean_nfl_df = clean_nfl_df.loc[(clean_nfl_df['Date'] > '2022-09-07') & (clean_nfl_df['Date'] < '2023-01-09')]
```

```python
# Splitting the Opp1 column to be able to one hot encode out later.
clean_nfl_df[['acr','Opp']] = clean_nfl_df['Opp1'].str.split(' ', expand=True)
clean_nfl_df
```

```python
# Don't need the date, Opp1
clean_nfl_df = clean_nfl_df[['Team','Pass_Yds','Pass_TD','Int','Rush_Yds','Rush_TD','Rec','Rec_Yds','Rec_TD','Tar','Opp']]
clean_nfl_df
```

```python
#Making a column for fantasy points
"""
Passing Yards: 1 point per 25 yards
Passing Touchdowns: 4 points
Passing Interceptions: -2 points
Rushing Yards: 1 point per 10 yards
Rushing Touchdowns: 6 points
Receptions: 1 points (only if using PPR scoring)
Receiving Yards: 1 point per 10 yards
Receiving Touchdowns: 6 points
2-Point Conversions: 2 points
Fumbles Lost: -2 points
Fumble Recovered for a Touchdown: 6 points
"""
clean_nfl_df['FP'] = (clean_nfl_df['Pass_Yds']/25) + (clean_nfl_df['Pass_TD']*4) - (clean_nfl_df['Int']*2) \
+ (clean_nfl_df['Rush_Yds']/10) + (clean_nfl_df['Rush_TD']*6) + (clean_nfl_df['Rec']) + (clean_nfl_df['Rec_Yds']/10)\
+(clean_nfl_df['Rec_TD']*6)
clean_nfl_df
```

```python
#Dropping Team and Opp to make a simple model
clean_nfl_df_no_opp_or_team = clean_nfl_df.drop(columns=['Team','Opp'])
clean_nfl_df_no_opp_or_team
```

```python
X = clean_nfl_df_no_opp_or_team.drop(columns='FP')
y = clean_nfl_df_no_opp_or_team['FP']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25,random_state=99)
```

```python
lin_model = LinearRegression()
```

```python
lin_model.fit(X_train,y_train)
```

```python
cross_val = cross_validate(lin_model,
                          X_train,
                          y_train,
                          scoring=('r2','neg_mean_squared_error','neg_root_mean_squared_error'),
                              return_train_score=True)
```

```python
cross_val_mse_scores = cross_val['test_neg_mean_squared_error']*-1
cross_val_rmse_scores = cross_val['test_neg_root_mean_squared_error']*-1
```

```python
# Extremely small scores, which means the model predicts very well. Although what it has to predict, FP, is formulaic...
# Next step is grid search to see if we can minimize even more.
print(cross_val_mse_scores)
print(cross_val_rmse_scores)
```

```python
# Time to try Ridge, Lasso, and Elasticnet
ridge_model = Ridge(random_state=99)
lasso_model = Lasso(random_state=99)
elnet_model = ElasticNet(random_state=99)
```

```python
ridge_model.fit(X_train,y_train)
lasso_model.fit(X_train,y_train)
elnet_model.fit(X_train,y_train)
```

```python
ridge_cross_val = cross_validate(ridge_model,
                          X_train,
                          y_train,
                          scoring=('r2','neg_mean_squared_error','neg_root_mean_squared_error'),
                              return_train_score=True)
```

```python
ridge_cross_val['test_neg_root_mean_squared_error']*-1
```

```python
lasso_cross_val['test_neg_root_mean_squared_error']*-1
```
